Complaints Table

```python
import pandas as pd

complaints_df = pd.read_csv('/content/complaints.csv')
print("Null values in each column of the complaints table:")
print(complaints_df.isnull().sum())
```

### Imputing Missing `compensation_amount` in `complaints_df`

We will categorize `resolution_days` and then use the median `compensation_amount` within each category to fill missing `compensation_amount` values.

```python

bins = [1, 5, 10, 15, complaints_df['resolution_days'].max() + 1]
labels = ['1-5 days', '5-10 days', '10-15 days', '15+ days']


complaints_df['resolution_days_category'] = pd.cut(
    complaints_df['resolution_days'],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True
)


median_compensation_by_category = complaints_df.groupby('resolution_days_category')['compensation_amount'].median()

print("Median compensation amount by resolution days category:")
print(median_compensation_by_category)


complaints_df['compensation_amount'] = complaints_df.groupby('resolution_days_category')['compensation_amount'].transform(lambda x: x.fillna(x.median()))


print("\nNull values in 'compensation_amount' after imputation:")
print(complaints_df['compensation_amount'].isnull().sum())


print("\nDataFrame after imputing missing compensation amounts:")
display(complaints_df.head())
```

```python
import os

output_folder = 'corrected_data'
output_filename_complaints = 'corrected_complaints.csv'


if not os.path.exists(output_folder):
    os.makedirs(output_folder)


output_filepath_complaints = os.path.join(output_folder, output_filename_complaints)


complaints_df.to_csv(output_filepath_complaints, index=False)

print(f"DataFrame successfully saved to '{output_filepath_complaints}'")
```

### Imputing Missing `preferred_channel` in `customers_df`

We will impute the missing `preferred_channel` based on two sources:
1.  **Primary Source (`orders_df`)**: For customers who have placed orders, we will use their most frequently used `booking_channel`.
2.  **Secondary Source (`app_events_df`)**: For customers who have not placed orders (and still have a null `preferred_channel`), we will check their `device_type` in `app_events_df`. If their `device_type` is 'Android' or 'iOS', we will set their `preferred_channel` to 'App'.

```python
import pandas as pd

orders_df = pd.read_csv('/content/orders.csv')


most_frequent_channel = orders_df.groupby('customer_id')['booking_channel'].agg(lambda x: x.mode()[0] if not x.mode().empty else None).reset_index()
most_frequent_channel.rename(columns={'booking_channel': 'inferred_preferred_channel'}, inplace=True)

customers_df = customers_df.merge(most_frequent_channel, on='customer_id', how='left')

customers_df['preferred_channel'] = customers_df['preferred_channel'].fillna(customers_df['inferred_preferred_channel'])

customers_df.drop(columns=['inferred_preferred_channel'], inplace=True)

print("Null values in 'preferred_channel' after primary imputation (from orders):")
print(customers_df['preferred_channel'].isnull().sum())
```

```python

customers_with_null_channel = customers_df[customers_df['preferred_channel'].isnull()]['customer_id'].unique()

app_events_for_null_customers = app_events_df[
    (app_events_df['customer_id'].isin(customers_with_null_channel)) &
    (app_events_df['device_type'].isin(['Android', 'iOS']))
].copy()

=inferred_app_channel = app_events_for_null_customers.groupby('customer_id')['device_type'].first().reset_index()
inferred_app_channel['inferred_app_channel'] = 'App'

customers_df = customers_df.merge(inferred_app_channel[['customer_id', 'inferred_app_channel']], on='customer_id', how='left')

customers_df['preferred_channel'] = customers_df['preferred_channel'].fillna(customers_df['inferred_app_channel'])

customers_df.drop(columns=['inferred_app_channel'], inplace=True)

print("\nNull values in 'preferred_channel' after secondary imputation (from app events):")
print(customers_df['preferred_channel'].isnull().sum())

print("\nDataFrame after imputing missing preferred channels:")
display(customers_df.head())
```

Deliveries Table




```python
import pandas as pd

deliveries_df = pd.read_csv('/content/deliveries.csv')

print("Null values in each column of the Deliveries table:")
print(deliveries_df.isnull().sum())
```

```python

deliveries_df['dispatch_time'] = pd.to_datetime(deliveries_df['dispatch_time'])
deliveries_df['delivery_completed_at'] = pd.to_datetime(deliveries_df['delivery_completed_at'])
app_events_df['event_timestamp'] = pd.to_datetime(app_events_df['event_timestamp'])

missing_delivery_completed_at_orders = deliveries_df[deliveries_df['delivery_completed_at'].isnull()]['order_id'].unique()


relevant_app_events = app_events_df[
    (app_events_df['order_id'].isin(missing_delivery_completed_at_orders)) &
    (app_events_df['event_type'].isin(['eta_refresh', 'track_order']))
].copy()


max_event_timestamp_for_delivery = relevant_app_events.groupby('order_id')['event_timestamp'].max().reset_index()
max_event_timestamp_for_delivery.rename(columns={'event_timestamp': 'inferred_delivery_completed_at'}, inplace=True)

deliveries_df = deliveries_df.merge(max_event_timestamp_for_delivery, on='order_id', how='left')

deliveries_df['delivery_completed_at'] = deliveries_df['delivery_completed_at'].fillna(deliveries_df['inferred_delivery_completed_at'])

deliveries_df.drop(columns=['inferred_delivery_completed_at'], inplace=True)

print("Null values in 'delivery_completed_at' after imputation from app_events:")
print(deliveries_df['delivery_completed_at'].isnull().sum())

print("\nDataFrame after imputing missing delivery completed times:")
display(deliveries_df.head())
```

```python

successful_deliveries = deliveries_df[deliveries_df['delivery_status'].isin(['OnTime', 'Delayed'])].copy()


successful_deliveries['trip_duration'] = successful_deliveries['delivery_completed_at'] - successful_deliveries['dispatch_time']


average_trip_duration = successful_deliveries['trip_duration'].mean()

print(f"Average trip duration for successful deliveries: {average_trip_duration}")

missing_delivery_completed_at_indices = deliveries_df[deliveries_df['delivery_completed_at'].isnull()].index

deliveries_df.loc[missing_delivery_completed_at_indices, 'delivery_completed_at'] = \
    deliveries_df.loc[missing_delivery_completed_at_indices, 'dispatch_time'] + average_trip_duration

print("\nNull values in 'delivery_completed_at' after final imputation:")
print(deliveries_df['delivery_completed_at'].isnull().sum())

print("\nDataFrame after final imputation of missing delivery completed times:")
display(deliveries_df.head())
```

```python
import os

output_folder = 'corrected_data'
output_filename_deliveries = 'corrected_deliveries.csv'


if not os.path.exists(output_folder):
    os.makedirs(output_folder)

output_filepath_deliveries = os.path.join(output_folder, output_filename_deliveries)

deliveries_df.to_csv(output_filepath_deliveries, index=False)

print(f"DataFrame successfully saved to '{output_filepath_deliveries}'")
```

```python

global_median_rating = deliveries_df['customer_rating_post_delivery'].median()
print(f"Global median customer rating: {global_median_rating:.2f}")

driver_avg_ratings = deliveries_df.groupby('driver_id')['customer_rating_post_delivery'].transform(lambda x: x.mean() if x.notnull().any() else None)

deliveries_df['customer_rating_post_delivery'] = deliveries_df['customer_rating_post_delivery'].fillna(driver_avg_ratings)

deliveries_df['customer_rating_post_delivery'] = deliveries_df['customer_rating_post_delivery'].fillna(global_median_rating)

print("\nNull values in 'customer_rating_post_delivery' after imputation:")
print(deliveries_df['customer_rating_post_delivery'].isnull().sum())

print("\nDataFrame after imputing missing customer ratings:")
display(deliveries_df.head())
```

```python
import pandas as pd

orders_df = pd.read_csv('/content/orders.csv')

print("Null values in each column of the Orders table:")
print(orders_df.isnull().sum())
```

### Imputing Missing `booking_channel` in `orders_df`

We will impute missing `booking_channel` values by:
1.  Identifying orders with a null `booking_channel`.
2.  Searching the `app_events_df` for records associated with these `order_id`s.
3.  Inferring the `booking_channel` based on the `device_type` in `app_events_df` (`Android`/`iOS` -> 'App', `Web` -> 'Web').

```python
orders_df['order_created_at'] = pd.to_datetime(orders_df['order_created_at'])
app_events_df['event_timestamp'] = pd.to_datetime(app_events_df['event_timestamp'])

orders_with_null_channel = orders_df[orders_df['booking_channel'].isnull()].copy()

null_order_ids = orders_with_null_channel['order_id'].unique()

# Filter app_events_df for these order_ids
relevant_app_events = app_events_df[app_events_df['order_id'].isin(null_order_ids)].copy()

# Create a mapping from order_id to inferred booking_channel based on device_type
def get_inferred_channel(device_types):
    if 'Android' in device_types or 'iOS' in device_types:
        return 'App'
    if 'Web' in device_types:
        return 'Web'
    return None

inferred_channels = relevant_app_events.groupby('order_id')['device_type'].apply(get_inferred_channel).reset_index(name='inferred_channel')

orders_df = orders_df.merge(inferred_channels, on='order_id', how='left')

orders_df['booking_channel'] = orders_df['booking_channel'].fillna(orders_df['inferred_channel'])

orders_df.drop(columns=['inferred_channel'], inplace=True)

print("\nNull values in 'booking_channel' after imputation:")
print(orders_df['booking_channel'].isnull().sum())

print("\nDataFrame after imputing missing booking channels:")
display(orders_df.head())
```

```python

orders_still_null_channel = orders_df[orders_df['booking_channel'].isnull()].copy()

merged_with_customers = orders_still_null_channel.merge(
    customers_df[['customer_id', 'preferred_channel']],
    on='customer_id',
    how='left',
    suffixes=('_order', '_customer')
)

orders_df.loc[orders_df['booking_channel'].isnull(), 'booking_channel'] = (
    orders_df.loc[orders_df['booking_channel'].isnull()].merge(
        customers_df[['customer_id', 'preferred_channel']],
        on='customer_id',
        how='left'
    )['preferred_channel']
)

print("\nNull values in 'booking_channel' after secondary imputation (from customers_df):")
print(orders_df['booking_channel'].isnull().sum())
```

```python

global_most_frequent_channel = orders_df['booking_channel'].mode()[0]
print(f"Global most frequent booking channel: {global_most_frequent_channel}")

orders_df['booking_channel'] = orders_df['booking_channel'].fillna(global_most_frequent_channel)

print("\nNull values in 'booking_channel' after final imputation (global mode):")
print(orders_df['booking_channel'].isnull().sum())

print("\nDataFrame after all booking channel imputations:")
display(orders_df.head())
```

```python
import pandas as pd


if 'orders_df' not in locals() and 'orders_df' not in globals():
    orders_df = pd.read_csv('/content/orders.csv')

if 'app_events_df' not in locals() and 'app_events_df' not in globals():
    app_events_df = pd.read_csv('/content/app_events.csv')

app_events_df['event_timestamp'] = pd.to_datetime(app_events_df['event_timestamp'])
orders_df['order_created_at'] = pd.to_datetime(orders_df['order_created_at'])


app_events_df = app_events_df.sort_values(by=['customer_id', 'event_timestamp']).reset_index(drop=True)
orders_df = orders_df.sort_values(by=['customer_id', 'order_created_at']).reset_index(drop=True)

app_events_to_merge = app_events_df[app_events_df['order_id'].isnull()].copy()

if not app_events_to_merge.empty:

    app_events_to_merge = app_events_to_merge.sort_values(by=['customer_id', 'event_timestamp']).reset_index(drop=True)


    merged_asof_result = pd.merge_asof(
        app_events_to_merge, # Use the freshly sorted subset
        orders_df[['customer_id', 'order_id', 'order_created_at']],
        left_on='event_timestamp',
        right_on='order_created_at',
        by='customer_id',
        direction='nearest', # Match nearest timestamp (before or after)
        tolerance=pd.Timedelta('1 hour'), # Within 1 hour
        suffixes=('_event', '_order_inferred')
    )


    inferred_ids = merged_asof_result[['event_id', 'order_id_order_inferred']].copy()
    inferred_ids.rename(columns={'order_id_order_inferred': 'inferred_order_id'}, inplace=True)


    app_events_df = app_events_df.merge(inferred_ids, on='event_id', how='left')


    app_events_df['order_id'] = app_events_df['order_id'].fillna(app_events_df['inferred_order_id'])


    app_events_df.drop(columns=['inferred_order_id'], inplace=True)
else:
    print("No null 'order_id' values remaining in app_events_df for time-based mapping with orders_df.")

print("Null values in 'order_id' after time-based mapping with orders_df:")
print(app_events_df['order_id'].isnull().sum())

print("\nDataFrame after time-based imputation:")
display(app_events_df.head())
```

```python
import pandas as pd

# Ensure deliveries_df is loaded before operations are performed on it
if 'deliveries_df' not in locals() and 'deliveries_df' not in globals():
    deliveries_df = pd.read_csv('/content/deliveries.csv')
# Ensure app_events_df is loaded as well, in case of partial execution
if 'app_events_df' not in locals() and 'app_events_df' not in globals():
    app_events_df = pd.read_csv('/content/app_events.csv')

# Ensure deliveries_df timestamp columns are in datetime format
deliveries_df['dispatch_time'] = pd.to_datetime(deliveries_df['dispatch_time'])
deliveries_df['delivery_completed_at'] = pd.to_datetime(deliveries_df['delivery_completed_at'])

# Filter app_events_df for remaining null order_ids and relevant event types
app_events_to_match = app_events_df[
    (app_events_df['order_id'].isnull()) &
    (app_events_df['event_type'].isin(['track_order', 'eta_refresh']))
].copy()

if not app_events_to_match.empty:
    # Prepare deliveries_df with relevant columns and ensure non-null timestamps
    delivery_tracking_info = deliveries_df[
        ['customer_id', 'order_id', 'dispatch_time', 'delivery_completed_at']
    ].dropna(subset=['dispatch_time', 'delivery_completed_at']).copy()

    # Merge app_events_to_match with delivery_tracking_info on customer_id
    merged_delivery_match = pd.merge(
        app_events_to_match,
        delivery_tracking_info,
        on='customer_id',
        how='left',
        suffixes=('_event', '_delivery')
    )

    # Filter for matches where event_timestamp is within dispatch_time and delivery_completed_at
    # And where the delivery order_id is not null (as it's the target)
    valid_matches = merged_delivery_match[
        (merged_delivery_match['event_timestamp'] >= merged_delivery_match['dispatch_time']) &
        (merged_delivery_match['event_timestamp'] <= merged_delivery_match['delivery_completed_at']) &
        (merged_delivery_match['order_id_delivery'].notnull())
    ].copy()

    if not valid_matches.empty:
        # From these valid matches, select the first inferred order_id for each event_id
        # This handles cases where an event might fall into multiple delivery windows
        # (e.g., if a customer has overlapping deliveries, pick the first one chronologically)
        inferred_delivery_order_ids = valid_matches.groupby('event_id').first()['order_id_delivery'].reset_index()
        inferred_delivery_order_ids.rename(columns={'order_id_delivery': 'inferred_order_id_from_delivery'}, inplace=True)

        # Merge these inferred order_ids back to the original app_events_df
        app_events_df = app_events_df.merge(inferred_delivery_order_ids, on='event_id', how='left')

        # Fill any remaining null 'order_id' values in app_events_df with the inferred ones from deliveries
        app_events_df['order_id'] = app_events_df['order_id'].fillna(app_events_df['inferred_order_id_from_delivery'])

        # Drop the temporary inferred column
        app_events_df.drop(columns=['inferred_order_id_from_delivery'], inplace=True)
    else:
        print("No valid matches found in deliveries_df for remaining null 'order_id' values.")
else:
    print("No null 'order_id' values remaining in app_events_df for tracking window imputation with deliveries_df.")


# Check remaining null values in 'order_id'
print("Null values in 'order_id' after tracking window imputation with deliveries_df:")
print(app_events_df['order_id'].isnull().sum())

# Display the first few rows to show the updated DataFrame
print("\nDataFrame after delivery tracking window imputation:")
display(app_events_df.head())
```
